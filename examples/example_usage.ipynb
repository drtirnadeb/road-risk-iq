# üß™ Example Usage ‚Äì RiskAnalyzer Package

This example demonstrates how to use the `RiskAnalyzer` Python package on a sample accidents dataset. The steps shown include:

- üì• Loading a small CSV file with accident records
- üó∫Ô∏è Computing state-level risk metrics (rate, severity, composite score)
- üìä Generating visualizations:
  - Choropleth map of composite risk
  - Accident trends by time of day
  - Distribution by weather condition

The goal is to validate that the package functions as intended before applying it to the full dataset.

---

**Sample States Included**:
- California (CA)
- Texas (TX)
- New York (NY)

**Input**:
- `sample_accidents.csv` (3 states, small subset for testing)
- Population dictionary with real 2020 census estimates

**Expected Outputs**:
- Printed state-level risk scores
- `risk_map.html` for interactive choropleth
- Matplotlib charts for time and weather patterns


```python
from accidents import RiskAnalyzer
print(RiskAnalyzer.__module__)
```

```python
# examples/example_usage.py

from accidents import RiskAnalyzer

# Define population dictionary
population_dict = {
    'CA': 39538223,
    'TX': 29145505,
    'NY': 20201249
}

# ‚úÖ Use CSV input
analyzer = RiskAnalyzer(
    data_path='../data/sample_accidents.csv',
    population_dict=population_dict
)

# Compute and print risk scores
state_stats = analyzer.compute_risk_scores()
print("\n‚úÖ Risk scores from CSV file:\n", state_stats)

# Plot results (will save/display plots)
analyzer.plot_risk_map()
analyzer.plot_time_analysis()
analyzer.plot_weather_analysis()
```
